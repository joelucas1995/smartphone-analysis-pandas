```python
import pandas as pd 
```

```python
# 1. Carregar o dataset
Lista_smartphones_abril= pd.read_csv("smartprix_smartphones_april_2026.csv")
```

```python
# 2. Descrição inicial dos dados
print("Informações gerais:")
print(Lista_smartphones_abril.info())
```

```python
print("\nValores nulos por coluna:")
print(Lista_smartphones_abril.isnull().sum())
```

```python
print("\nValores únicos por coluna:")
print(Lista_smartphones_abril.nunique())
```

```python
# 3. Tratamento de qualidade dos dados
# Remover colunas com valores constantes
Lista_smartphones_abril = Lista_smartphones_abril.loc[:, Lista_smartphones_abril.nunique() > 1]
```

```python
# Remover linhas com valores nulos
Lista_smartphones_abril = Lista_smartphones_abril.dropna()
```

```python
# Conferir novamente nulos
print("\nApós tratamento, nulos por coluna:")
print(Lista_smartphones_abril.isnull().sum())
```

```python
# 4. Perguntas e análises

# Pergunta 1: Qual a média de preço por categoria?
print("\nMédia de preço por categoria:")
print(Lista_smartphones_abril.groupby("price_category")["price"].mean())
```

```python
# Pergunta 2: Qual marca tem mais modelos listados?
print("\nMarca com mais modelos:")
print(Lista_smartphones_abril["brand_name"].value_counts().head())
```

```python
# Pergunta 3: Qual sistema operacional mais aparece?
print("\nSistema operacional mais frequente:")
print(Lista_smartphones_abril["os"].value_counts().head())
```

```python
# Pergunta 4: Qual a média de capacidade de bateria por categoria de preço?
print("\nMédia de bateria por categoria de preço:")
print(Lista_smartphones_abril.groupby("price_category")["battery_capacity(mAh)"].mean())
```

```python
# Pergunta 5: Quais modelos possuem mais de 3 câmeras traseiras?
print("\nModelos com mais de 3 câmeras traseiras:")
print(Lista_smartphones_abril[Lista_smartphones_abril["rear_camera_count"] > 3][["brand_name","model","rear_camera_count"]])
```
